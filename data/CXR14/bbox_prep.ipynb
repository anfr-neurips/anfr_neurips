```python
# open each image in the first 10 rows and display the image and the bounding box
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import numpy as np

for i in range(len(bbox)-25, len(bbox)):
    im = np.array(Image.open("images/" + bbox["Image Index"][i]))
    fig,ax = plt.subplots(1)
    ax.imshow(im, cmap='gray')
    # add the index to the title
    plt.title("Image " + str(i))
    width = bbox["w"][i]
    height = bbox['h]'][i]
    x = bbox['Bbox [x'][i]
    y = bbox["y"][i]
    rect = patches.Rectangle((x,y),width,height,linewidth=1,edgecolor='r',facecolor='none')

    ax.add_patch(rect)
    plt.show()
```

```python
# resize both the images and the bounding boxes to 224x224
import os
from PIL import Image
import pandas as pd
import numpy as np

# load the bbox list
bbox = pd.read_csv("BBox_List_2017.csv")

# resize the images
for i in range(len(bbox)):
    im = Image.open("images/" + bbox["Image Index"][i])
    im = im.resize((224,224))
    im.save("resized_images/" + bbox["Image Index"][i])

# resize the bounding boxes
bbox["x"] = bbox["x"] * (224/1024)
bbox["y"] = bbox["y"] * (224/1024)
bbox["w"] = bbox["w"] * (224/1024)
bbox["h"] = bbox["h"] * (224/1024)

```
